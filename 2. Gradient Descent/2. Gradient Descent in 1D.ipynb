```python
import numpy as np
import matplotlib.pyplot as plt

from IPython import display
display.set_matplotlib_formats('svg')
#this draws figure in vector format instead of pixel format
#SVG = Scalable Vector Graphics
```

```python
#Gradient Descent in 1D

def fx(x):
    return 3*x**2- 3*x+ 4

def deriv(x):
    return 6*x - 3
```

```python
#plot the function and its derivative

#define a range for x
x = np.linspace(-2,2,2001)
#This will create an array of 2001 evenly spaces values between -2 and 2.
#plotting
plt.plot(x,fx(x),x,deriv(x))
plt.xlim(x[[0,-1]])
#x[[0,-1]] selects the first and last element in the array.
#plt.xlim(x[[0,-1]]) , This sets the limit of the x-axis in a 
#matplotlib plot based on first and last elements of array x.

plt.grid()
#It is used to create a grid for the plot.

plt.xlabel('x')
plt.ylabel('f(x)')

plt.legend(['y','dy'])
#It is used to create a legend .
plt.show()
```

```python
#random starting point
localmin = np.random.choice(x,1)
print(localmin)
#This will randomly choose one element from the array x.

#Learning parameters
learning_rate = .01
training_epochs = 100

#implementation of gradient descent algorithm
#run through training and store all results

for i in range(training_epochs):
    grad = deriv(localmin)
    localmin = localmin - learning_rate * grad

localmin
```

```python
#plot the results

plt.plot(x,fx(x),x,deriv(x))
plt.plot(localmin,deriv(localmin),'ro')
plt.plot(localmin,fx(localmin),'ro')

plt.xlim(x[[0,-1]])
plt.grid()
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend(['f(x)','df','f(x) min']) 
plt.title('Empirical local minimum: %s'%localmin[0])
plt.show()    
```

```python
import numpy as np

# Assuming training_epochs is defined as the number of training epochs
training_epochs = 1000  # Example value
learning_rate = 0.001  # Example learning rate
modelparams = np.zeros((training_epochs, 2))

# Initialize localmin (assuming it's defined before the loop starts)
localmin = np.random.choice(x,1)

for i in range(training_epochs):
    grad = deriv(localmin)
    localmin = localmin - learning_rate * grad
    
    # Store localmin and grad in the ith row of modelparams
    modelparams[i, 0] = localmin  # Assign localmin to the first column
    modelparams[i, 1] = grad  # Assign grad to the second column
```

```python
#plot the gradient over iterations

fig,ax = plt.subplots(1,2,figsize=(12,4))

'''
plt.subplots(1, 2, figsize=(12, 4)): This function creates a figure (fig) and a pair of axes (ax) arranged in a single row (1) and two columns (2), resulting in two subplots side by side.
figsize=(12, 4): Specifies the width and height of the figure in inches. In this case, it's 12 inches wide and 4 inches tall.
'''

for i in range(2):
    ax[i].plot(modelparams[:,1],'o-')
    ax[i].set_xlabel('Iteration')
    ax[i].set_title(f'final estimated minimum : {localmin[0]:5f}')
    
ax[0].set_ylabel('local minimum')
ax[1].set_ylabel('Derivative')

plt.show()
```

```python
'''If you have very small learning rate, you need to have more iterations
beacuse then the model is going to learn more slowly. The learning rate 
of 0.001 and 1000 training epochs not necessarily going to be the optimal
set of parameters for every single model and for every single metaparameter 
and every single dataset.

In fact the optimal lerning rate depends on the flavour of gradient descent
algorithm and backgropagation, how deep the model is and other factors.
'''
```

```python
'''
Question: Repeat the gradient descent to find the minimum of the 
following function:

f(x) = cos(2πx) + x^2 ,  x ∈ {-2,2}
'''
```

```python
import numpy as np
import matplotlib.pyplot as plt

from IPython import display
display.set_matplotlib_formats('svg')
```

```python
def fx(x):
    return np.cos(2*np.pi*x) + x**2

def deriv(x):
    return -np.sin(2*np.pi*x)*2* np.pi + 2*x
```

```python
x=np.linspace(-2,2,2001)
plt.plot(x,fx(x),x,deriv(x))

plt.xlim(x[[0,-1]])

plt.xlabel('x')
plt.ylabel('fx')

plt.grid()
plt.legend(['y','dy'])
```

```python
local_minimum = np.random.choice(x,1)

learning_rate=.01
training_epochs=100

for i in range(training_epochs):
    grad = deriv(local_minimum)
    local_minimum = local_minimum - learning_rate * grad
print(local_minimum)
```

```python
#plot the results
plt.plot(x,fx(x),x,deriv(x))
plt.plot(local_minimum,deriv(local_minimum),'ro')
plt.plot(local_minimum,fx(local_minimum),'ro')

plt.xlim(x[[0,-1]])
plt.grid()

plt.xlabel('x')
plt.ylabel('f(x)')
    
plt.legend(['f(x)','df','f(x) min'])
plt.show()
```

```python
'''
Here, we can observe that it gives us local_maximum instead of local_minimum.
This is because the function gives 0 at 0. And since;
local_minimum = local_minimum - learning_rate * grad
local_minimum will equal local_minimum all time, and therefore there
is no learning at all. This case is called vanishing descent.
Usually, In practise the probability of this happeining is minimal.
'''
```

```python
local_minimum = np.random.choice(x,1)
learning_rate=.01
learning_epochs=100
modelparams=np.zeros((training_epochs,2))

for i in range(training_epochs):
    grad = deriv(local_minimum)
    local_minimum = local_minimum - learning_rate * grad
    
    modelparams[i,0]=local_minimum
    modelparams[i,1]=grad
```

```python
fig, ax = plt.subplots(1,2,figsize=[12,4])
for i in range(2):
    ax[i].plot(modelparams[:,1],'o-')
    ax[i].set_xlabel('Iteration')
    ax[i].set_title(f'final estimated minimum : {local_minimum[0]:5f}')
    
ax[0].set_ylabel('local minimum')
ax[1].set_ylabel('Derivative')

plt.show()
```
