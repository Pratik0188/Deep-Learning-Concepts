```python
#Softmax
import numpy as np
import torch
import torch.nn as nn
import matplotlib.pyplot as plt
```

```python
#manually in numpy
z=np.random.randint(-5,high=15,size=25)
print(z)

#the list of numbers
#z = [1,2,3]

#compute the softmax result
num = np.exp(z)
den = np.sum(np.exp(z))
sigma = num/den

#compare
plt.plot(z,sigma,'ko')
plt.xlabel('Original Number (z)')
plt.ylabel('Softmaxified $\sigma$')
plt.yscale('log')
plt.title('$\sum\sigma$ = %g' %np.sum(sigma))
plt.show()
```

```python
#Using pytorch
#create an instance of the softmax activation class
softfun= nn.Softmax(dim=0)

#then apply the data to that function
sigmaT=softfun(torch.Tensor(z))

#now we get the results
print(sigmaT)
```

```python
#show that they are same
plt.plot(sigma,sigmaT,'ko')
plt.xlabel('"Manual" softmax')
plt.ylabel('Pytorch nn.Softmax')
plt.title(f'The two methods co-relate at r={np.corrcoef(sigma,sigmaT)[0,1]}')
plt.show()
```

```python
#probability of an event happening
p=.25

#Not the correct formulae:
H = -(p*np.log(p))
print('Wrong Entropy:'+ str(H))

#Here, this is just the probability of an event happening. There is also the probability 
#of an event not happening and the probability of it happening is q=1-p=.75 .Then, we would
#need to sum over all of the values of X not only the one value of x.
```

```python
#the correct way to compute entropy
x=[.25,.75]

H=0
for p in x:
    H+= -p*np.log(p)
    
print('Correct entropy:' + str(H))
```

```python
#also correct, written out for N=2 events.

#Binary cross entropy
H = -(p*np.log(p) + (1-p)*np.log(1-p))
print('Correct entropy:' + str(H))
```

```python
#Cross-Entropy
p=[1,0] #sum=1
q=[.25,.75] #sum=1

H=0
for i in range(len(p)):
    H-= p[i]*np.log(q[i])

print('Cross Entropy:'+str(H))
```

```python
#also correct, written out for N=2 events.
H = -(p[0]*np.log(q[0]) + q[1]*np.log(q[1]))
print('Correct Entropy:' + str(H))

#simplification
H = -np.log(q[0])
print('Manually simplifies:' + str(H))
```

```python
#argmax
#Using numpy

#create a vector
v= np.array([1,40,2,-3])

#find and report the maximum and minimum values
minval = np.min(v)
maxval = np.max(v)

print('Min,max: %g%g' %(minval,maxval))

#now for the argmin/max
minidx = np.argmin(v)
maxidx = np.argmax(v)

print('Min,max indices: %g%g' %(minidx,maxidx)), print(' ')
```

```python
#repeat with matrix
M = np.array([[0,1,10],
             [20,8,5]])
print(M), print(' ')

#various minima in this matrix!
minvals1 = np.min(M)
minvals2 = np.min(M,axis=0)
minvals3 = np.min(M,axis=1)

#print them out
print(minvals1)
print(minvals2)
print(minvals3)
```

```python
#Using pytorch
v = torch.tensor([1,40,2,-3])

#find and report the maximum and minimum values:
minval = torch.min(v)
maxval = torch.max(v)

print('Min,max: %g%g' %(minval,maxval))

#now for argmin/max
minidx = torch.argmin(v)
maxidx = torch.argmax(v)

print('Min,max indices: %g%g' %(minidx,maxidx)), print(' ')

#confirm
print(f' Min val is {v[minidx]}, max val is {v[maxidx]}')
```

```python
#Mean 
x = [1,2,4,6,5,4,0]
n = len(x)

#compute the mean
mean1 = np.mean(x)
mean2 = np.sum(x) / n

#print them
print(mean1)
print(mean2)
```

```python
#Variance
var1 = np.var(x)
var2 = (1/(n-1))* np.sum((x-mean1)**2)
#here, if you divide by n, its a biased measured of variance. if you divide by n-1 it is 
#unbiased measure of variance.

print(var1)
print(var2)
```

```python
var3 = np.var(x,ddof=1)
#ddof = degree of freedom
print(var3)
print(var2)
```

```python
#does it matter for large N?

N=1000
x = np.random.randint(0,high=20,size=N)

var0 = np.var(x,ddof=0) #default
var1 = np.var(x,ddof=1) #unbiased

print(var0)
print(var1)
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#create a list of numbers to compute the mean and variance of
x=[1,2,4,6,5,0,-4,5,-2,6,10,-9,1,3,3-6]
n = len(x)

#compute the population mean
popmean = np.mean(x)

#compute a sample mean
sample = np.random.choice(x,size=5,replace=True)
sampmean = np.mean(sample)

#print them
print(popmean)
print(sampmean)
```

```python
#compute lots of sample means

#number of experiments to run
nExpers =10000

#run the experiment!
sampleMeans = np.zeros(nExpers)
for i in range(nExpers):
    
    #step1 : draw a sample
    sample = np.random.choice(x,size=5,replace=True)
    
    #step2: compute its mean
    sampleMeans[i] = np.mean(sample)
    
#show the results in a histogram
plt.hist(sampleMeans,bins=40,density=True)
plt.plot([popmean,popmean],[0,3],'m--')
plt.ylabel('Count')
plt.xlabel('Sample mean')
plt.show()
```

```python
#t-test using scipy

import numpy as np
import matplotlib.pyplot as plt

import scipy.stats as stats
```

```python
#parameters
n1 = 30  #samples in dataset1
n2 = 40  #sample in dataset2
mu1 = 1  #population mean in dataset1
mu2 = 2  #population mean in dataset2

#generate the data
data1 = mu1 + np.random.randn(n1)
data2 = mu2 + np.random.randn(n2)

#plot them
plt.plot(np.zeros(n1),data1,'ro',markerfacecolor='w',markersize=14)
plt.plot(np.ones(n2),data2,'bs',markerfacecolor='w',markersize=14)
plt.xlim([-1,2])
plt.xticks([0,1],labels=['Group1','Group2'])
plt.show()

#here , in fig , the t-test is telling us that Group2 has significant statistically 
#significant larger mean tha. group1.
```

```python
#t-test via stats package

#_ind = independent samples
t,p = stats.ttest_ind(data1,data2)
print(t)
print(p)
```

```python
#common wat to show t-test results in a plot:
fig = plt.figure(figsize=(10,4))
plt.rcParams.update({'font.size':12}) #change the font size

#here 15 is just a random no, we generated based on visualization.
plt.plot(0+np.random.randn(n1)/15, data1, 'ro', markerfacecolor='w',markersize=14)
plt.plot(1+np.random.randn(n2)/15, data2, 'bs', markerfacecolor='w',markersize=14)
plt.xlim([-1,2])
plt.xticks([0,1],labels=['Group1','Group2'])

#set the title to include the t-value and p-value
plt.title(f't = {t:.2f}, p={p:.3f}')
plt.show()
```

```python
#Derivative using sympy library
import sympy as sym
import sympy.plotting.plot as symplot
```

```python
#creating symbolic variables in sympy
x = sym.symbols('x')

#create a function
fx = 2*x**2

#compute its derivative
df = sym.diff(fx,x)

#print them
print(fx)
print(df)
```

```python
#plot them
symplot(fx,(x,-4,4),title='The function')
plt.show()

symplot(df,(x,-4,4),title='Its derivative')
plt.show()
```

```python
#repeat with relu and sigmoid

#create symbolic functions
relu=sym.Max(0,x)
sigmoid= 1/(1+sym.exp(-x))

#graph the functions
p=symplot(relu,(x,-4,4),label='ReLU',show=False,line_color='blue')
p.extend(symplot(sigmoid,(x,-4,4),label='Sigmoid',show=False,line_color='red'))
p.legend = True
p.title = 'The Functions'
p.show()

#graph the derivatives
p = symplot(relu,(x,-4,4),label='ReLU',show=False,line_color='blue')
p.extend(symplot(sigmoid,(x,-4,4),label='Sigmoid',show=False,line_color='red'))
p.legend = True
p.title = 'The functions'
p.show()
```

```python
#chain rule and product rule
import numpy as np
import sympy as sym

#make the equations look nicer
from IPython.display import display

#this library will make the equations look nices and latex formatted.
```

```python
#create symbolic variables in sympy
x = sym.symbols('x')

#create two functions
fx = 2*x**2
gx = 4*x**3 - 3*x**4

#compute their individual derivatives
df = sym.diff(fx)
dg = sym.diff(gx)

#apply the product rule 'manually'
manual = df*gx + fx*dg
thewrongway = df*dg

#via sympy
viasympy = sym.diff(fx*gx)

#print everything
print('The functions:')
display(fx)
display(gx)
print(' ')

print('Their derivatives')
display(df)
display(dg)
print(' ')

print('Manual product rule:')
display(manual)
print(' ')

print('Via sympy:')
display(viasympy)
print(' ')

print('The wrong way:')
display(thewrongway)
```
