```python
import numpy as np
import torch
import torch.nn as nn

import matplotlib.pyplot as plt
from IPython import display
display.set_matplotlib_formats('svg')
```

```python
#create data

nPerClust = 100
blur = 1

A = [1,1]
B = [5,1]
#These are just the center co-ordinates in (1,1) and (5,1)

#generate data
a = [A[0] + np.random.randn(nPerClust)*blur, A[1]+np.random.randn(nPerClust)*blur]
b = [B[0]+ np.random.randn(nPerClust)*blur, B[1]+np.random.randn(nPerClust)*blur]
#Adding noises to the center points

#true labels
labels_np = np.vstack((np.zeros((nPerClust,1)),np.ones((nPerClust,1))))

#concatenate into a matrix
data_np = np.hstack((a,b)).T

#convert to a pytorch tensor
data = torch.tensor(data_np).float()
labels = torch.tensor(labels_np).float()

#show the data
fig = plt.figure(figsize=(5,5))
plt.plot(data[np.where(labels==0)[0],0],data[np.where(labels==0)[0],1],'bs')
plt.plot(data[np.where(labels==1)[0],0],data[np.where(labels==1)[0],1],'ko')
plt.title('The binary classification')
plt.xlabel('Dimension 1')
plt.ylabel('Dimension 2')
plt.show()
```

```python
#Inspecting data types
print(type(data_np))
print(np.shape(data_np))
print(' ')

print(type(data))
print(np.shape(data))
```

```python
#build the model
ANNclassify = nn.Sequential(
    nn.Linear(2,1),    # input layer
    nn.ReLU(),         # activation unit
    nn.Linear(1,1),     # output unit
    nn.Sigmoid(),      # final activation unit
    )

ANNclassify
```

```python
#other model features

learningRate = 0.01

#loss function
lossfun = nn.BCELoss()
#Note: You'll learn in the 'Metaparameters' section that it's better 
#to use BCEwithLogitsLoss, but it is fine for now.
#BCE= Binary Cross Entropy
#optimizer
optimizer = torch.optim.SGD(ANNclassify.parameters(),lr=learningRate)
```
